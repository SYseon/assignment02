```python
import numpy as np
import matplotlib.pyplot as plt
import random
from scipy.misc import derivative
```

# Define my function:   $ f(x) = \frac{1}{1+e^{-x}} $

```python
def my_func(x):
    result = 1/(1+np.exp(-x))
    return result
```

# Define my deravative function:  $ f(x) = \frac{e^{-x}}{(1+e^{-x})^{2}} $

```python
def deravative_func(x):
    return derivative(my_func, x, dx = 1e-5)
```

# Define taylor approxiamtion form.

```python
def taylor_approximation(domain, point):
    result = deravative_func(point) * (domain - point) + my_func(point)
    return result
```

# Define domain and ouput of above functions.

```python
x=np.arange(-5,5,0.1)
y=my_func(x)
d_y=deravative_func(x)
```

# Randomly choose 3 points.

```python
point = np.zeros(3)

for i in range(3):
    point[i] = random.choice(x)
    print(point[i])
```

# Compute the graphs

```python
plt.figure(1)
plt.plot(x, y, 'b', label="function")
plt.plot(x,taylor_approximation(x,point[0]), 'g', label="taylor approximation on fist point")
plt.plot(x,taylor_approximation(x,point[1]), 'y', label="taylor approximation on second point")
plt.plot(x,taylor_approximation(x,point[2]), 'c', label="taylor approximation on third point")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
